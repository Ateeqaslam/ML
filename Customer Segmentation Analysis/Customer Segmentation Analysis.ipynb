```python
# columns we used for clustering
selected_columns = ['Income','Kidhome','MntWines','MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds','Age','NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases','MntTotal']
x = Data[selected_columns]
```

```python
# scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Determing the number of clusters using Elbow Method
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  k_means = KMeans(n_clusters = i, random_state=42)
  # print(k_means)
  k_means.fit(x_scaled)
  wcss.append(k_means.inertia_)
  # print(wcss)
```

```python
# plot the elbow
plt.figure(figsize=(10,4))
plt.plot(range(1,11),wcss,marker = 'o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Square)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

As we got the first Elbow at 3 and the second one at 6. We will focus on 3 clusters

```python
# applying the kmeans cluster on 3
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(x_scaled)

# adding cluster label to the Data
Data['Cluster'] = kmeans.labels_
```

```python
print(Data[['Income','Age','MntTotal','Cluster']].head())
```

```python
#  Visualize the clusters
plt.figure(figsize= (10,4))
plt.scatter(x_scaled[:,13],x_scaled[:,0],c= Data['Cluster'],cmap= 'viridis')
plt.title('Clusters (3)')
plt.xlabel('MntTotal')
plt.ylabel('Income')
plt.show()
```
