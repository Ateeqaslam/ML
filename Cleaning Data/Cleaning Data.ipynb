```python
Data.dropna(axis= 0,inplace= True)
print(Data.isnull().sum())
```

```python
Data.duplicated().sum()
```

```python
Data.head()
```

```python
Data = Data[Data['availability_365'] <= 365]
```

```python
Data['price'].describe()
```

```python
q1 = Data['price'].quantile(0.5)
q3 = Data['price'].quantile(0.75)
print(q1)
print(q3)
```

```python
IQR = q3 - q1
Threshold = 1.5
lowerband = q1 - (Threshold * IQR)
upperband = q3 + (Threshold * IQR)
print(lowerband)
print(upperband)
```

```python
Data['price'].max()
```

```python
Data = Data[(Data['price'] >= lowerband) & (Data['price'] <= upperband)]
```
